## Определение функций

```python
# Проверка нулевой гипотезы с помощью теста Стьюдента.
def get_stat_dif(column):
    # отбираются все строки, относящиеся к больш. рейтингу
    cols = students.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(students.loc[students.loc[:, column] == comb[0], 'score'],
                     students.loc[students.loc[:, column] == comb[1], 'score']).pvalue \
                <= 0.05/len(combinations_all):  # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break


# Границы выбросов
def get_outliners(column):
    median = students[column].median()
    quan25 = students[column].quantile(0.25)
    quan75 = students[column].quantile(0.75)
    IQR = quan75 - quan25
    left = quan25 - 1.5*IQR,
    right = quan75 + 1.5*IQR
    print(f"Границы выбросов для столбца '{column}': [{left}, {right}]")
```

### Замена пропусков на моду

```python
cols = students.columns.values
for col in cols:
    if str(students[col].dtype) == 'category':
        number = students[col].isnull().sum()
        if number > 0:
            item_mode = students[col].describe()['top']
            students[col] = students[col].fillna(item_mode)
            print(f"{number} значений заменено в '{col}'")
```
